# **General cleaning & organizing**

```python
time_points = ["Week 1", "Week 2", "Week 3", "Week 4", "Week 5", "Week 6"]

def calculate_group_averages(groups):
    group_averages = {"height": [], "leaves": [], "buds": []}

    for metric in ["height", "leaves", "buds"]:
        for group in groups:
            avg_values = np.mean([plant_data[plant][metric] for plant in group], axis=0)
            group_averages[metric].append(avg_values)

    return group_averages

plant_groups = [
    ["Plant One", "Plant Two", "Plant Three"],
    ["Plant Four", "Plant Five", "Plant Six"],
    ["Plant Seven", "Plant Eight", "Plant Nine"]
]

group_averages = calculate_group_averages(plant_groups)
```

```python
group_colors = ["green", "red", "orange"]
group_names = ["Control", "Treatment Two", "Treatment One"]
```

```python
for metric in ["height", "leaves", "buds"]:
    plt.figure(figsize=(10, 6))

    lines = []
    for i, (group_avg, group_name, group_color) in enumerate(zip(group_averages[metric], group_names, group_colors)):
        line, = plt.plot(time_points, group_avg, marker='o', linestyle='-', color=group_color, label=group_name)
        lines.append(line)

    ylabel = f"{metric.capitalize()} (cm)" if metric == "height" else f"{metric.capitalize()}" # units addition

    plt.title(f"Average {metric.capitalize()} by Week for Plant Groups")
    plt.ylabel(ylabel)
    plt.legend(handles=[lines[0], lines[1], lines[2]])  # Set legend order explicitly
    plt.grid(True)
    plt.show()
```
